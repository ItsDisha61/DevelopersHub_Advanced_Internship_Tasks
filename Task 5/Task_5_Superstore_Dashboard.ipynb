# TASK 5: Superstore Interactive Business Dashboard

```python
# Import Libraries
import streamlit as st
import pandas as pd
import plotly.express as px
```

```python
# Load Dataset
df = pd.read_csv(r"C:\Task 5\superstore.csv")
```

```python
# Clean Dataset
df.drop_duplicates(inplace=True)

df.dropna(inplace=True)
```

```python
# Dashboard Title
st.set_page_config(
    page_title="Superstore Dashboard",
    layout="wide"
)

st.title("📊 Interactive Business Dashboard")
```

```python
# Sidebar Filters
region = st.sidebar.selectbox(
    "Region",
    ["All"] + sorted(df["Region"].unique())
)

category = st.sidebar.selectbox(
    "Category",
    ["All"] + sorted(df["Category"].unique())
)

sub_category = st.sidebar.selectbox(
    "Sub-Category",
    ["All"] + sorted(df["Sub.Category"].unique())
)
```

```python
# Apply Filters
filtered_df = df.copy()

if region != "All":
    filtered_df = filtered_df[
        filtered_df["Region"] == region
    ]

if category != "All":
    filtered_df = filtered_df[
        filtered_df["Category"] == category
    ]

if sub_category != "All":
    filtered_df = filtered_df[
        filtered_df["Sub.Category"] == sub_category
    ]
```

```python
# Key Performance Indicators (KPIs)
total_sales = filtered_df["Sales"].sum()

total_profit = filtered_df["Profit"].sum()

top_customer = (
    filtered_df.groupby("Customer.Name")["Sales"]
    .sum()
    .idxmax()
)

col1, col2, col3 = st.columns(3)

col1.metric(
    "Total Sales",
    f"${total_sales:,.2f}"
)

col2.metric(
    "Total Profit",
    f"${total_profit:,.2f}"
)

col3.metric(
    "Top Customer",
    top_customer
)
```

```python
# Sales by Region
sales_region = (
    filtered_df.groupby("Region")["Sales"]
    .sum()
    .reset_index()
)

fig = px.bar(
    sales_region,
    x="Region",
    y="Sales",
    color="Region",
    title="Sales by Region"
)

st.plotly_chart(fig, use_container_width=True)
```

```python
# Profit by Region
profit_region = (
    filtered_df.groupby("Region")["Profit"]
    .sum()
    .reset_index()
)

fig = px.bar(
    profit_region,
    x="Region",
    y="Profit",
    color="Region",
    title="Profit by Region"
)

st.plotly_chart(fig, use_container_width=True)
```

```python
# Sales by Category
sales_category = (
    filtered_df.groupby("Category")["Sales"]
    .sum()
    .reset_index()
)

fig = px.pie(
    sales_category,
    names="Category",
    values="Sales",
    title="Sales by Category"
)

st.plotly_chart(fig, use_container_width=True)
```

```python
# Top 5 Customers by Sales
top5 = (
    filtered_df.groupby("Customer.Name")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

fig = px.bar(
    top5,
    x="Customer.Name",
    y="Sales",
    color="Sales",
    title="Top 5 Customers by Sales"
)

st.plotly_chart(fig, use_container_width=True)
```

```python
# Display Filtered Dataset
st.subheader("Filtered Dataset")

st.dataframe(filtered_df)
```
